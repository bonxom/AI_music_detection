```python
from pathlib import Path
import shutil
from tqdm import tqdm

# Folder sau khi bạn unzip Kaggle
SRC_ROOT = Path("../archive/fma_medium")

# Folder đích chứa tất cả real audio
REAL_DIR = Path("../archive/real_songs")
REAL_DIR.mkdir(parents=True, exist_ok=True)

mp3_files = sorted(SRC_ROOT.rglob("*.mp3"))

print("Found mp3:", len(mp3_files))

copied = 0
skipped = 0

for src in tqdm(mp3_files):
    dst = REAL_DIR / src.name

    if dst.exists():
        skipped += 1
        continue

    shutil.move(str(src), str(dst))
    copied += 1

print("Copied:", copied)
print("Skipped existing:", skipped)
print("Total in real:", len(list(REAL_DIR.glob('*.mp3'))))
print("Real folder:", REAL_DIR)
```
