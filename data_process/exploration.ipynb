# Download real video

```python
from pathlib import Path
import re
import shutil

try:
    import yt_dlp
    from yt_dlp.utils import DownloadError
except ImportError as e:
    raise ImportError('Chưa cài yt-dlp. Chạy: pip install -U "yt-dlp[default]"') from e


OUTPUT_DIR = Path("archive/real_songs")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

LIMIT = None
START_FROM = "real_02995"

FFMPEG_LOCATION = shutil.which("ffmpeg")
NODE_LOCATION = shutil.which("node")
DENO_LOCATION = shutil.which("deno")
BUN_LOCATION = shutil.which("bun")
QJS_LOCATION = shutil.which("qjs") or shutil.which("quickjs")


def sanitize_filename(name: str, max_len: int = 180) -> str:
    name = str(name).strip()
    name = re.sub(r'[\\/:*?"<>|]', "_", name)
    name = re.sub(r"\s+", " ", name).strip(" .")
    if not name:
        name = "untitled"
    return name[:max_len]


def choose_unique_stem(output_dir: Path, stem: str, youtube_id: str) -> str:
    if not list(output_dir.glob(f"{stem}.*")):
        return stem

    alt_stem = f"{stem}_{youtube_id}"
    if not list(output_dir.glob(f"{alt_stem}.*")):
        return alt_stem

    return ""


def build_js_runtimes():
    if DENO_LOCATION:
        return {"deno": {"path": DENO_LOCATION}}
    if NODE_LOCATION:
        return {"node": {"path": NODE_LOCATION}}
    if BUN_LOCATION:
        return {"bun": {"path": BUN_LOCATION}}
    if QJS_LOCATION:
        return {"quickjs": {"path": QJS_LOCATION}}
    return None


def build_ydl_opts(outtmpl: str):
    if not FFMPEG_LOCATION:
        raise RuntimeError(
            "Muốn xuất MP3 thì cần ffmpeg/ffprobe. "
            "Cài ffmpeg rồi chạy lại: sudo apt install ffmpeg"
        )

    js_runtimes = build_js_runtimes()

    ydl_opts = {
        "quiet": True,
        "noprogress": True,
        "retries": 5,
        "fragment_retries": 5,
        "concurrent_fragment_downloads": 4,
        "noplaylist": True,
        "outtmpl": outtmpl,
        "windowsfilenames": False,
        "overwrites": False,
        "format": "bestaudio/best",
        "postprocessors": [{
            "key": "FFmpegExtractAudio",
            "preferredcodec": "mp3",
            "preferredquality": "192",
        }],
        "ffmpeg_location": FFMPEG_LOCATION,
    }

    if js_runtimes is not None:
        ydl_opts["js_runtimes"] = js_runtimes

    return ydl_opts, js_runtimes


def classify_error(msg: str) -> str:
    s = msg.lower()
    if (
        "video unavailable" in s
        or "this video is not available" in s
        or "private video" in s
        or "unavailable video" in s
    ):
        return "unavailable"
    if "sign in to confirm your age" in s or "age-restricted" in s:
        return "age_restricted"
    if "ffmpeg is not installed" in s:
        return "missing_ffmpeg"
    if "no supported javascript runtime could be found" in s:
        return "missing_js_runtime"
    return "other"


def download_audio(url: str, outtmpl: str):
    ydl_opts, js_runtimes = build_ydl_opts(outtmpl)
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(url, download=True)
    return {
        "js_runtimes": js_runtimes,
        "info": info,
    }


name_col = "file_name" if "file_name" in df_real.columns else "filename"
if name_col not in df_real.columns:
    raise ValueError("Không tìm thấy cột tên file: cần file_name hoặc filename")
if "youtube_id" not in df_real.columns:
    raise ValueError("Không tìm thấy cột youtube_id trong df_real")

work_df = df_real[["youtube_id", name_col]].dropna().copy()
work_df["youtube_id"] = work_df["youtube_id"].astype(str).str.strip()
work_df[name_col] = work_df[name_col].astype(str).str.strip()
work_df = work_df[
    (work_df["youtube_id"] != "") & (work_df[name_col] != "")
].drop_duplicates("youtube_id")

# resume từ START_FROM nếu có
if START_FROM:
    mask = work_df[name_col].astype(str).eq(START_FROM)
    if mask.any():
        start_idx = mask.idxmax()
        work_df = work_df.loc[start_idx:]

if LIMIT is not None:
    work_df = work_df.head(LIMIT)

ok_count = 0
skip_count = 0
unavailable_count = 0
failures = []

env_logged = False

for idx, row in work_df.reset_index(drop=True).iterrows():
    youtube_id = row["youtube_id"]
    target_name = sanitize_filename(row[name_col])
    url = f"https://www.youtube.com/watch?v={youtube_id}"

    final_stem = choose_unique_stem(OUTPUT_DIR, target_name, youtube_id)
    if not final_stem:
        skip_count += 1
        continue

    outtmpl = str(OUTPUT_DIR / f"{final_stem}.%(ext)s")

    try:
        result = download_audio(url, outtmpl)

        if not env_logged:
            js_runtimes = result["js_runtimes"]
            print(
                f"Env: ffmpeg=OK, "
                f"js_runtime={list(js_runtimes.keys())[0] if js_runtimes else 'NONE'}"
            )
            if not js_runtimes:
                print("Cảnh báo: không tìm thấy JS runtime hỗ trợ (deno/node/bun/quickjs).")
            env_logged = True

        ok_count += 1
        if ok_count % 1000 == 0:
            print(f"Đã tải xong {ok_count} mẫu.")

    except DownloadError as e:
        msg = str(e)
        err_type = classify_error(msg)
        failures.append({
            "youtube_id": youtube_id,
            "name": final_stem,
            "error_type": err_type,
            "error": msg,
        })
        if err_type == "unavailable":
            unavailable_count += 1

    except Exception as e:
        failures.append({
            "youtube_id": youtube_id,
            "name": final_stem,
            "error_type": "other",
            "error": str(e),
        })

print("\n===== SUMMARY =====")
print(f"Tổng: {len(work_df)}")
print(f"Thành công: {ok_count}")
print(f"Bỏ qua (đã có file): {skip_count}")
print(f"Unavailable/private/removed: {unavailable_count}")
print(f"Lỗi khác: {len(failures) - unavailable_count}")

if failures:
    print("\nMột vài lỗi đầu tiên:")
    for item in failures[:10]:
        print(item)
```
