```python
columns = ["filename", "path", "fake_type", "label"]
```

```python
def get_index_df(len_clip):
    if len_clip not in [5, 120]:
        raise ValueError("Độ dài clip chỉ có 5s hoặc 120s")
    
    in_PATH = ""
    output_file = f"../crop_data/crop{len_clip}.csv"
    
    rows = []

    if len_clip == 5:
        in_PATH = "../crop_data/crop5"
    else:
        in_PATH = "../crop_data/crop120"

    # Real-type process
    real_PATH = os.path.join(in_PATH, "real")
    for filename in os.listdir(real_PATH):
        file_PATH = os.path.join(real_PATH, filename)
        rows.append([filename, file_PATH[3:], "Real", 1])

    # Fake-type process
    fake_PATH = os.path.join(in_PATH, "fake")
    for filename in os.listdir(fake_PATH):
        match = df_fake.loc[df_fake["filename"] + ".wav" == filename, "label"]
        fake_class = match.iloc[0] if not match.empty else None
        file_PATH = os.path.join(fake_PATH[3:], filename)
        rows.append([filename, file_PATH, fake_class, 0])

    df = pd.DataFrame(rows, columns=columns)
    df.to_csv(output_file, index=False)
    print(f"Đã tạo csv index cho crop {len_clip}s")
    return df

        
```

```python
get_index_df(5)
get_index_df(120)
```
